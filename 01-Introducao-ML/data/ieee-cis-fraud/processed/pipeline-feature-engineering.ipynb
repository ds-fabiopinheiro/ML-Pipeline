```python
# Conteúdo do notebook (resumido)
import pandas as pd
from pathlib import Path

# Load data
train_trans = pd.read_csv('../raw/train_transaction.csv')
train_id = pd.read_csv('../raw/train_identity.csv')

# Merge tables
df = pd.merge(train_trans, train_id, on='TransactionID', how='left')

# Feature engineering
df['hour'] = df['TransactionDT'] % (24*60*60) // (60*60)
df['avg_transaction_by_card'] = df.groupby('card1')['TransactionAmt'].transform('mean')

# Save to Parquet
df.to_parquet('../processed/dataset_enriquecido.parquet', index=False)
```
